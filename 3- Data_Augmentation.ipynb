## SINGLE IMAGE DATA AUGMENTATION

```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#Load the image file, resizing it to 200x200 pixels (required by our model)
img = image.load_img('3- train/shoes.jpg', target_size=(200, 200))
```

```python
img
```

```python
type(img)
```

- Remember its not a numpy array so you don't use any numpy array function on it:

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='nearest') #nearest, constant, reflect, wrap
```

```python
img = image.img_to_array(img)  # Convert the image to numpy array
```

```python
type(img)
```

```python
input_batch = img.reshape((1, 200, 200, 3))  # Reshape the image to (1, 200, 200, 3), as the model expects, 3 is for RGB channels
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='aug_image'): #datagen.flow means work on only one image
    i += 1
    if i == 10:
        break

```
